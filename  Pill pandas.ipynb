## Reload the dataset so that null values are correctly interpreted (repeat section 4, delete columns).

```python
dfcopy.columns
```

```python
columns_to_load=['Name', 'City', 'Country', 'IATA', 'ICAO', 'Timezone', 'DST', 'TZ',
       'Type', 'source']
```

```python
dfn = pd.read_csv("airports.csv", na_values= ["\\N"], sep=";",usecols= columns_to_load)
dfn
```

## Checks the entire dataframe for null values.

```python
dfn.isnull()
```

```python
dfn.isna().sum()
```

Now, the percentage on nulls per column:

```python
dfn.isna().sum()/len(dfn)*100
```

## Overwrites the null values of the IATA and ICAO columns with the value 'UNKNOWN'

```python
dfn.ICAO.fillna("UNKNOWN", inplace=True)
```

```python
dfn.ICAO.value_counts()
```

## Changes the type of the DST and TZ variables to categorical.

```python
dfn["DST"] = pd.Categorical(df["DST"])
dfn["TZ"] = pd.Categorical(df["TZ"])
dfn.info()
```

## Obtain a statistical summary of the categorical variables.

```python
dfn.describe(include="category")
```

## Groups the dataframe by airport type, showing the type count.

```python
dfn.groupby("Type").count()
```

```python
dfn.groupby("Type").max()
```

## Select the name of the cities whose airport type is "port"

```python
dfn[(dfn.Type == "port")] [["Name", "City", "Type"]]
```

## Shows all the rows of the fields name of the airport, name of the country and, name of the city, whose country is Spain.

```python
dfn.columns
```

```python
dfn[(dfn["Country"] == "Spain")] [["Name", "Country", "City"]]
```

## Shows the name of the country and the airport belonging to the city of Madrid and Barcelona. Are all the records from Spain?

```python
dfn[(dfn.City=="Madrid") | (dfn.City =="Barcelona")] [["Country","Name"]]
```

## Save the previous results in a csv called Madrid_Barcelona.csv

```python
dfMaBa=dfn[(dfn.City=="Madrid") | (dfn.City =="Barcelona")] [["Country","Name"]]
dfMaBa.to_csv("Madrid Barcelona Airports")
```
